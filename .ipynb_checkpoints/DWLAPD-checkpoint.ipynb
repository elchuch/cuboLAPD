# Data Warehousing

### Carga de modulos

```python
import pandas as pd
import numpy as np
from datetime import datetime

```

### Lectura de datos

```python
path="/home/vega97/Documentos/lapd.csv"
df=pd.read_csv(path)
df
```

### Preparación de datos

```python
df.info()# información acerca de los datos
```

```python
(df.isnull().sum()/df.shape[0])*100 # porcentaje de missing data por features 
```

```python
df['DATE OCC']# cambiar el valor de la fecha,quitando el formato  'fecha hora' dejando  fecha
```

##### Cambiando el formato de la variables tipo fecha

```python
df['Date Rptd']=df['Date Rptd'].str.replace(' \d{2}:\d{2}:\d{2} [AP]M', '')# q
```

```python
df['DATE OCC']=df['DATE OCC'].str.replace(' \d{2}:\d{2}:\d{2} [AP]M', '')
```

```python
df['DATE OCC'] =pd.to_datetime(df['DATE OCC'],format="%m/%d/%Y")
df['DATE OCC']=df['DATE OCC'].dt.strftime("%Y%m%d")
```

```python
df.head()
```

### Creando tabla de hechos 

```python
df_Th=df[['DATE OCC','AREA','AREA NAME','Vict Age','Vict Descent','Vict Sex','Crm Cd Desc','Premis Desc','Status Desc','LAT','LON']].groupby(['DATE OCC','AREA','AREA NAME','Vict Descent','Vict Sex','Crm Cd Desc','Premis Desc','Status Desc','LAT','LON']).agg(['max','min'])
df_Th
```

```python
df_Th.columns = ["_".join(x) for x in df_Th.columns]
```

```python
df_Th=df_Th.reset_index(drop=False)
df_Th.head()
```

```python
df_Th.rename(columns={'Vict Age_max':'age'},inplace=True)#cambiando el nombre de una columna por edad
```

```python
l=df_Th.columns.tolist()
l
lNew=['date_occ','area','area_name','vict_descent','vict_sex','crm_cd_desc','premis_dec','status_desc','lat','lon','age']
for i in range(len(l)):
    df_Th.rename(columns={l[i]:lNew[i]},inplace=True)
df_Th.head()
```

```python
#df_Th.drop('level_0', axis=1,inplace=True)
#df_Th.drop('index', axis=1,inplace=True)
df_Th.head()
```

### Tabla de dimension fecha

```python
date=df['DATE OCC']
dataFechas=date.drop_duplicates()
dataFechas=dataFechas.reset_index(drop=True)
dataFechas=dataFechas.sort_values().reset_index(drop=True)
dataFechas
```

```python
td_fecha=dataFechas.to_frame()
td_fecha
```

```python
td_fecha['date']=pd.to_datetime(td_fecha['DATE OCC'])
td_fecha['month']=td_fecha['date'].dt.strftime("%B")
td_fecha['year']=td_fecha['date'].dt.year
td_fecha.rename(columns={'DATE OCC':'date_occ'},inplace=True)

td_fecha
```

### Preservación de tabla de hecho y dimensión


```python
df_Th.to_csv("th_lapd.csv",index=False)

```

```python
td_fecha.to_csv("td_fecha.csv",index=False)
```
