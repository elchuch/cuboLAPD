```python
with open("/lapd_oltp.json", "r") as f:
    cred = js.load(f)
```

```python
cred
```

```python
host = cred['host']
port = cred['port']
user = cred['user']
pwd = cred['pwd']
db = cred['db']
```

```python
flavor = "mysql"
client = "pymysql"
```

```python
conn = create_engine(
    f"{flavor}+{client}://{user}:{pwd}@{host}:{port}/{db}").connect()
```

```python
df = pd.read_sql(con=conn, sql="Show tables")
```

```python
df
```

```python
datoltp.to_sql(name="LAPD", con=conn, index=False, if_exists="replace")
```
